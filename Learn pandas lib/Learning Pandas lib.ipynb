# New Data Frame

```python
df = pd.DataFrame({
    "name": ['Alfred','Batman','Alfred'],
    "toy" : [np.nan, np.nan,np.nan],
    "born" : [pd.NaT, pd.Timestamp("2021-05-28"), pd.NaT]
})
```

```python
df.head()
```

```python
# This will drop a row which contain NaN
df.dropna()
```

```python
# to remove the column which contain NaN
df.dropna(how="all", axis=1)
```

```python
# to drop duplicates
df.drop_duplicates(subset = ["name"])
```

```python
df.drop_duplicates(subset=["name"], keep = "last")
```

```python
df.drop_duplicates(subset=["name"], keep = False)
```

```python
# to find the shape
df.shape
```

```python
# to get info about data
df.info()
```

```python
# to know the value count
df['name'].value_counts()
```
